```python
# filling the missing value for the continous variables for proper data visualization

def fillNA(column,value):
    if value == 'mean':
        data[column].fillna(data[column].mean(), inplace = True)
    elif value == 'median':
        data[column].fillna(data[column].median(), inplace = True)
    else:
        data[column].fillna(value, inplace = True)
        
col_val = [['ShortPassing','mean'],['Volleys','mean'],['Dribbling','mean'],['Curve','mean'],['FKAccuracy','mean'],
           ['LongPassing','mean'],['BallControl','mean'],['HeadingAccuracy','mean'],['Finishing','mean'],['Crossing','mean'],
           ['Weight','200lbs'],['Contract Valid Until',2019],['Height',"5'11"],['Loaned From','None'],['Joined','Jul 1, 2018'],
           ['Jersey Number',8],['Body Type','Normal'],['Position','ST'],['Club','No Club'],['Work Rate','Medium/ Medium'],
           ['Skill Moves','median'],['Weak Foot',3],['Preferred Foot','Right'],['International Reputation',1],
           ['Wage','€200K']]

for col,val in col_val:
    fillNA(col,val)

data.fillna(0, inplace = True)
```

>It looks like the entries doesn't belong to the same category,so i think that the best way to deal with such column is to replace M and K with 1e6 and 1e3 respectively.

```python
def conv2float(entry):
    entry = entry.replace('€', '')
    if 'M' in entry:
        return float(entry.replace('M', ''))*1000000
    elif 'K' in entry:
        return float(entry.replace('K', ''))*1000
    
data['Value'] = data['Value'].apply(lambda x: conv2float(x))
data['Wage'] = data['Wage'].apply(lambda x: conv2float(x))
```
